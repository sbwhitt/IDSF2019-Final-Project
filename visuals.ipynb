```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Model performances after merging genres
```

```python
names = ('Random Forest', 'Naive Bayes', 'K Nearest Neighbor', 'Decision Tree')
acc = [0.7006887777485303, 0.32971395981398616, 0.5199613933491269, 0.5821268754935509]
f1 = [0.65, 0.36, 0.52, 0.58]
```

```python
fig, ax = plt.subplots()
index = np.arange(len(names))
bar_width = 0.35
opacity = 0.8

b1 = plt.bar(index, acc, bar_width,
alpha=opacity,
color='#5287fa',
label='Accuracy')

b2 = plt.bar(index + bar_width, f1, bar_width,
alpha=opacity,
color='#45e6a8',
label='F1 Score')

plt.xlabel('Model')
plt.ylabel('Accuracy/F1 Score')
plt.title('Model Performance (Score)')
plt.xticks(index + bar_width, names)
plt.legend()
plt.tight_layout()
```

```python
names = ('Random Forest', 'Naive Bayes', 'K Nearest Neighbor', 'Decision Tree')
time = [637.330695152, 12.6477518082, 152.774731874, 322.526809931]

b = plt.bar(index + bar_width, time, bar_width,
alpha=opacity,
color='#f05459',
label='Time')

plt.xlabel('Model')
plt.ylabel('Time (seconds)')
plt.title('Model Performance (Time)')
plt.xticks(index + bar_width, names)
plt.legend()
plt.tight_layout()
```

```python
# Sentiment Analysis
```

```python
df = pd.read_pickle("./data.pkl")
```

```python
fig, ax = plt.subplots(figsize=(15,7))
index = np.arange(len(df.groupby(['genre'])))
bar_width = 0.35
opacity = 0.8

genres = ['Country/Folk', 'Electronic', 'Hip-Hop/R&B', 'Jazz', 'Pop', 'Rock/Metal']
pos = df.groupby(['genre']).mean()['pos_score']
neg = df.groupby(['genre']).mean()['neg_score']

b1 = plt.bar(index, pos, bar_width,
alpha=opacity,
color='#45e6a8',
label='Positive Score')

b2 = plt.bar(index + bar_width, neg, bar_width,
alpha=opacity,
color='#f05459',
label='Negative Score')

plt.xlabel('Genre')
plt.ylabel('Score')
plt.title('Sentiment Analysis')
plt.xticks(index + bar_width, genres)
plt.legend(loc="upper right")
plt.axis()
axes = plt.gca()
ax.set_ylim([0.06, 0.18])
plt.tight_layout()
```
